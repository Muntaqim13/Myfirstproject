```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.cluster import KMeans
```

```python
Customers=pd.read_csv('Mall_Customers.csv')
Customers
```

PLOTTING THE DATA

```python
plt.scatter(Customers['Age'],Customers['Annual Income (k$)'])
plt.title('MALL_CUSTOMERS')
plt.xlabel('Age')
plt.ylabel('Annual income')
plt.show()
```

SELECTING THE FEATURES(Age, Annual Income and Spending Score)

```python
Cluster=Customers.iloc[:,2:]
Cluster
```

INITIAL CLUSTERING USING 4 CLUSTERS.

```python
mall=KMeans(4)
Initial=mall.fit(Cluster)
```

GETTING THE CLUSTERING RESULTS

```python
results=Initial.fit_predict(Cluster)
results
```

ADDING THE RESULT TO THE NORMAL TABLE

```python
Customers['Clustering result']=results
Customers
```

PLOTTING THE DATA

```python
plt.scatter(Customers['Age'],Customers['Annual Income (k$)'],c=Customers['Clustering result'],cmap='rainbow')
plt.title('cluster analysis')
plt.xlabel('Age')
plt.ylabel('Annual Income')
plt.show()
```

DETERMINING THE APPROPRIATE NUMBER OF CLUSTERS USING THE ELBOW METHOD.

```python
#WCSS of the initial cluster
Initial.inertia_
```

```python
#iteration
WCSS=[]

for i in range(1,11):
     kmeans=KMeans(i)
     kmeans.fit(Cluster)
     WCSS_iteration=kmeans.inertia_
     WCSS.append(WCSS_iteration)
```

```python
WCSS
```

CREATING THE ELBOW PLOT TO CHOOSE THE APPROPRITE NUMBER OF CLUSTERS.

```python
#plotting the number_of_clusters against WCSS
number_of_clusters=range(1,11)
plt.plot(number_of_clusters,WCSS)
```

USING 5 CLUSTERS

```python
#from the elbow plot, we deduced that using 3 or 5 cluster is better than using 4
new_cluster=KMeans(5)
new=new_cluster.fit(Cluster)
new
```

RESULT OF THE NEW CLUSTER

```python
new_result=new.fit_predict(Cluster)
new_result
```

ADDING THE NEW RESULT TO THE TABLE

```python
Customers['new_results']=new_result
Customers
```

PLOTTING WITH THE NEW RESULT

```python
plt.scatter(Customers['Age'],Customers['Annual Income (k$)'],c=Customers['new_results'],cmap='rainbow')
plt.title('FINAL CLUSTER')
plt.xlabel('Age')
plt.ylabel('Annual income')
plt.show()
```
